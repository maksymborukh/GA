```python
import numpy as np
from math import e, ceil, log, sin, inf, pi
import random
```

```python
#f = lambda x1, x2: e**sin(x1**2)+x2
f = lambda x: x*sin(10*pi*x)+1
```

```python
a_x1 = -2
b_x1 = 2

a_x2 = 0
b_x2 = 1
```

```python
N = 100
N_t = 5
p_m = 0.1
p_c = 0.8
max_t = 30
max_t_nc = 10
h = 1e-2
```

```python
a = -1
b = 2
```

```python
find_power_of_2 = lambda a, b, h: ceil( log( (b-a)/h, 2 ) )
```

```python
def rand_ind(p):
    ind = ""
    for i in range(p):
        temp = str(random.randint(0, 1))
        ind += temp 
    return(ind)
```

```python
n = find_power_of_2(a, b, h)
```

```python
def M(x, a, b, n):
    x_r = x[::-1]
    k = (b-a)/(2**n-1)
    s = 0
    for i in range(n):
        s += 2**i * int(x_r[i])
    return a + k*s
```

```python
def eval_x(x, a, b, n):
    return f(M(x, a, b, n))
```

```python
def eval_P(P, a, b, n, N):
    best_ind = 0
    res = 0
    eval_P = inf
    for i in range(N):
        temp = M(P[i], a, b, n)
        if f(temp) < eval_P:
            res = temp
            best_ind = P[i]
            eval_P = f(temp)
    return eval_P, best_ind, res
```

```python
eval_P(['11011', '01011', '00000', '00011'], a, b, 5, 4)
```

```python
def tournament(P, a, b, n, N, N_t):
    s = []
    r = []
    for i in range(N_t):
        r.append( random.randint(0, N-1) )  
        s.append( eval_x(P[r[i]], a, b, n) )
    return r[s.index(min(s))]
```

```python
tournament(['11011', '01011', '10001', '00011'], a, b, 5, 4, 2)
```

```python
def mutation(x, p_m, n):
    x = x[::-1]
    ind = ''
    if random.uniform(0, 1) < p_m:
        r = random.randint(0, n)
        for i in range(n):           
            if i == r:
                ind += str(abs(int(x[r])-1))
            else:
                ind += x[i] 
    else:
        ind = x
    return ind[::-1]
```

```python
def cross(x1, x2, p_c, n):
    x1 = x1[::-1]
    x2 = x2[::-1]
    ind1 = ''
    ind2 = ''
    if random.uniform(0, 1) < p_c:
        r = random.randint(1, n)
        for i in range(n):           
            if i <= r:
                ind1 += x2[i]
                ind2 += x1[i]
            else:
                ind1 += x1[i]
                ind2 += x2[i]
    else:
        ind1 = x1
        ind2 = x2
    return ind1[::-1], ind2[::-1]
```

```python
cross('11011', '00011', p_c, 5)
```

```python
P_t = []
t = 0
P_t.append([])
for i in range(N):
    P_t[t].append(rand_ind(n))
res = eval_P(P_t[t], a, b, n, N)
t_m = 0
for i in range(1, max_t):
    if t_m > max_t_nc:
        break;
    
    if eval_P(P_t[t], a, b, n, N)[0] < res[0]:
        t_m = 0
        res = eval_P(P_t[t], a, b, n, N)
    else:
        t_m += 1

    t += 1
    P_t.append([])
    while len(P_t[t]) < N:
        ind_tour = []
        for j in range(2):
            ind_tour.append(tournament(P_t[t-1], a, b, n, N, N_t))
        ind_m = []
        for j in range(2):
            ind_m.append(mutation(P_t[t-1][ind_tour[j]], p_m, n))
            
        ind_c = cross(ind_m[0], ind_m[1], p_c, n)
        
        P_t[t].append(ind_c[0])
        P_t[t].append(ind_c[1])
```
